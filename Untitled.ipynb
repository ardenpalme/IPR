### Helper Functions

````python
def send_graphql_query_to_subgraph(api_key, 
                                   query, 
                                   subgraph_url="https://gateway.thegraph.com/api/subgraphs/id",
                                   url_ext="5zvR82QoaXYFyDEKLZ9t6v9adgnptxYpKpSbxtgVENFV",
                                   variables = None):
  headers = {
    'Content-Type': 'application/json',
    'Authorization': f'Bearer {api_key}'
  }

  # Prepare the request payload
  payload = {'query': query}
  if variables:
      payload['variables'] = variables
  
  # Send the GraphQL request to the Subgraph
  response = post(f"{subgraph_url}/{url_ext}", headers=headers, json=payload)

  # Check if the request was successful
  if response.status_code == 200:
      return response.json()
  else:
      print("Error:", response.text)
      return None
````

### Fees Estimates

The code below is a subgraph API query to retrieve the 
```
{
  poolDayDatas(
    first: 10, 
    orderBy: date, 
    where: {
      pool: "0x1d42064fc4beb5f8aaf85f4617ae8b3b5b8bd801", 
      date_gt: 1633642435
    }
  ) {
    date
    liquidity  // in range liquidity at end of period
    sqrtPrice  // current price tracker at end of period
    token0Price  // price of token0 (X) - derived from sqrtPrice
    token1Price  // price of token1 (Y) - derived from sqrtPrice
    feeGrowthGlobal0X128 
    feeGrowthGlobal1X128 
  }
}
```

$\newcommand{\X}{\mathcal{X}}$
$\newcommand{\Y}{\mathcal{Y}}$

$\Phi^\X_R\equiv$`feeGrowthGlobal0X128`

$\Phi^\Y_R\equiv$`feeGrowthGlobal1X128`

feeGrowthInside = Cumulative fees generated in pool ÷ Active liquidity at that time

sqrt_price is an integer stored as Q64.96 fixed-point format.

USDC has 6 'decimals'
WETH has 18 'decimals' 

Q128.128 fixed-point for fees

````python
N=1000
WETH_USDC_pool_addr="0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
query1 = f"""{{
  poolDayDatas(
    first: {N}, 
    orderBy: date, 
    where: {{
      pool: "{WETH_USDC_pool_addr}", 
      date_gt: {int(start_dt.timestamp())}
    }}
  ) {{
    date
    liquidity
    tick
    feeGrowthGlobal0X128
    feeGrowthGlobal1X128
  }}
}}
"""
print(query1)
````

````python
raw_data = send_graphql_query_to_subgraph(api_keys['SUBGRAPH_API_KEY'], query1)
````

````python
rows = []
for ele in raw_data["data"]["poolDayDatas"]:
    rows.append({
        'date': datetime.fromtimestamp(ele['date'], UTC),  
        'tick': float(ele['tick']),                        # current tick at end of period
        'active_liq': float(ele['liquidity']),             # in range liquidity at end of period
        'acc_fees_X': float(ele['feeGrowthGlobal0X128']),  # globally-accumulated fees in pool 
        'acc_fees_Y': float(ele['feeGrowthGlobal1X128']),
    })
df = pd.DataFrame(rows)
````

````python
getcontext().prec = 50

beta_p = Decimal(1.0001)
WETH_dec = 18
USDC_dec = 6
````

````python
df = df.set_index('date')
df['price'] = 1/((beta_p**df['tick'].apply(Decimal)) / Decimal(10**(WETH_dec-USDC_dec)))
df['sqrt_price']=np.sqrt(df['price'])
````

````python
# convert fees from Q128 format to denominated in respective token 
divisor = Decimal(2) ** 128

df['acc_fees_X'] = df['acc_fees_X'].astype("str").apply(Decimal) / divisor
df['acc_fees_Y'] = df['acc_fees_Y'].astype("str").apply(Decimal) / divisor
df['fees_X_hat'] = df['active_liq'].apply(Decimal) * df['acc_fees_X'].diff()
df['fees_Y_hat'] = df['active_liq'].apply(Decimal) * df['acc_fees_Y'].diff()
````

````python
fig, ax = plt.subplots(1,1,figsize=(8,4))
sns.lineplot(data=df, x="price", y="fees_X_hat", ax=ax)
ax.set_yscale('log')
plt.show()
````

````python
fig, ax = plt.subplots(1,1,figsize=(8,4))
sns.lineplot(data=df, x="date", y="fees_Y_hat", ax=ax)
ax.set_yscale('log')
plt.show()
````
