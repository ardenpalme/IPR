````python
start_date_str = "2026-01-01"
start_dt = datetime.strptime(start_date_str, "%Y-%m-%d").replace(tzinfo=UTC)
print(int(start_dt.timestamp()))
````

````python
def send_graphql_query_to_subgraph(api_key, 
                                   query, 
                                   subgraph_url="https://gateway.thegraph.com/api/subgraphs/id/5zvR82QoaXYFyDEKLZ9t6v9adgnptxYpKpSbxtgVENFV",
                                   variables = None):
  headers = {
    'Content-Type': 'application/json',
    'Authorization': f'Bearer {api_key}'
  }

  # Prepare the request payload
  payload = {'query': query}
  if variables:
      payload['variables'] = variables
  
  # Send the GraphQL request to the Subgraph
  response = post(subgraph_url, headers=headers, json=payload)

  # Check if the request was successful
  if response.status_code == 200:
      return response.json()
  else:
      print("Error:", response.text)
      return None
````

````python
# retrieve API keys from environment
api_keys = dict()
with open(".env") as f:
    for line in f.readlines():
        elems = line.split("=")
        api_keys[elems[0].strip()] = elems[1].strip()
api_keys
````

````python
query1 = f"""{{
  poolDayDatas(
    first: 10, 
    orderBy: date, 
    where: {{
      pool: "0x1d42064fc4beb5f8aaf85f4617ae8b3b5b8bd801", 
      date_gt: {int(start_dt.timestamp())}
    }}
  ) {{
    date
    liquidity
    sqrtPrice
    feeGrowthGlobal0X128
    feeGrowthGlobal1X128
    token0Price
    token1Price
    
  }}
}}
"""
````

````python
raw_data = send_graphql_query_to_subgraph(api_keys['SUBGRAPH_API_KEY'], query1)
# data = json.loads(raw_data)
````

````python
rows = []
for ele in raw_data["data"]["poolDayDatas"]:
    rows.append({
        'date': datetime.fromtimestamp(ele['date'], UTC),
        'sqrt_price': float(ele['sqrtPrice']),
        'active_liq': float(ele['liquidity']),
        'fees_X': float(ele['feeGrowthGlobal0X128']),
        'fees_Y': float(ele['feeGrowthGlobal1X128'])
    })
df = pd.DataFrame(rows)
````

````python
df = df.set_index('date')
````

````python
df['fees_X'] = df['fees_X'] / df['active_liq'] 
df['fees_Y'] = df['fees_Y'] / df['active_liq']
````

````python
sns.lineplot(data=df, x='sqrt_price', y='fees_X')
````
