### Helper Functions

````python
def send_graphql_query_to_subgraph(api_key, 
                                   query, 
                                   subgraph_url="https://gateway.thegraph.com/api/subgraphs/id",
                                   url_ext="5zvR82QoaXYFyDEKLZ9t6v9adgnptxYpKpSbxtgVENFV",
                                   variables = None):
  headers = {
    'Content-Type': 'application/json',
    'Authorization': f'Bearer {api_key}'
  }

  # Prepare the request payload
  payload = {'query': query}
  if variables:
      payload['variables'] = variables
  
  # Send the GraphQL request to the Subgraph
  response = post(f"{subgraph_url}/{url_ext}", headers=headers, json=payload)

  # Check if the request was successful
  if response.status_code == 200:
      return response.json()
  else:
      print("Error:", response.text)
      return None
````

### Fees Estimates

The code below is a subgraph API query to retrieve the 
```
{
  poolDayDatas(
    first: 10, 
    orderBy: date, 
    where: {
      pool: "0x1d42064fc4beb5f8aaf85f4617ae8b3b5b8bd801", 
      date_gt: 1633642435
    }
  ) {
    date
    liquidity  // in range liquidity at end of period
    sqrtPrice  // current price tracker at end of period
    token0Price  // price of token0 (X) - derived from sqrtPrice
    token1Price  // price of token1 (Y) - derived from sqrtPrice
    feeGrowthGlobal0X128 
    feeGrowthGlobal1X128 
  }
}
```

$\newcommand{\X}{\mathcal{X}}$
$\newcommand{\Y}{\mathcal{Y}}$

$\Phi^\X_R\equiv$`feeGrowthGlobal0X128`

$\Phi^\Y_R\equiv$`feeGrowthGlobal1X128`

feeGrowthInside = Cumulative fees generated in pool ÷ Active liquidity at that time

sqrt_price is an integer stored as Q64.96 fixed-point format.

USDC has 6 'decimals'
WETH has 18 'decimals' 

Q128.128 fixed-point for fees

````python
lookback_period = relativedelta(weeks=1)
END_DATE = datetime.today().strftime("%Y-%m-%d")
START_DATE = (datetime.today() - lookback_period).strftime("%Y-%m-%d")
````

````python
start_dt = datetime.strptime(START_DATE, "%Y-%m-%d").replace(tzinfo=UTC)
print(int(start_dt.timestamp()))
````

````python
N=lookback_period.days + 1
WETH_USDC_pool_addr="0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
query1 = f"""{{
  poolDayDatas(
    first: {N}, 
    orderBy: date, 
    where: {{
      pool: "{WETH_USDC_pool_addr}", 
      date_gt: {int(start_dt.timestamp())}
    }}
  ) {{
    date
    liquidity
    tick
    feeGrowthGlobal0X128
    feeGrowthGlobal1X128
    feesUSD
  }}
}}
"""
print(query1)
````

````python
raw_data = send_graphql_query_to_subgraph(api_keys['SUBGRAPH_API_KEY'], query1)
````

````python
rows = []
for ele in raw_data["data"]["poolDayDatas"]:
    rows.append({
        'date': datetime.fromtimestamp(ele['date'], UTC),  
        'tick': float(ele['tick']),                        # current tick at end of period
        'active_liq': float(ele['liquidity']),             # in range liquidity at end of period
        'acc_fees_X': float(ele['feeGrowthGlobal0X128']),  # globally-accumulated fees in pool 
        'acc_fees_Y': float(ele['feeGrowthGlobal1X128']),
        'fees_USD': float(ele['feesUSD']),
    })
df = pd.DataFrame(rows)
````

````python
getcontext().prec = 50

beta_p = Decimal(1.0001)
WETH_dec = 18
USDC_dec = 6
Q128 = Decimal(2) ** 128
````

````python
df = df.set_index('date')
df['price'] = 1/((beta_p**df['tick'].apply(Decimal)) / Decimal(10**(WETH_dec-USDC_dec)))
df['sqrt_price']=np.sqrt(df['price'])
````

````python
# convert fees from Q128 format to denominated in respective token 
df['acc_fees_X'] = df['acc_fees_X'].astype("str").apply(Decimal) / Q128
df['acc_fees_Y'] = df['acc_fees_Y'].astype("str").apply(Decimal) / Q128
````

````python
def getFeeGrowthOutside(tau: int):
    query = f"""
    {{
      ticks(where: {{ 
        	pool: "{WETH_USDC_pool_addr}", 
        	tickIdx: {tau} 
      }}) {{
        feeGrowthOutside0X128
        feeGrowthOutside1X128
      }}
    }}
    """
    raw_data = send_graphql_query_to_subgraph(api_keys['SUBGRAPH_API_KEY'], query)
    if len(raw_data["data"]["ticks"]) == 0:
        return 0,0
    feeGrowthOutside0 = Decimal(raw_data["data"]["ticks"][0]["feeGrowthOutside0X128"]) / Q128
    feeGrowthOutside1 = Decimal(raw_data["data"]["ticks"][0]["feeGrowthOutside1X128"]) / Q128
    print(f"{tau} : {feeGrowthOutside0}, {feeGrowthOutside1}")
    return feeGrowthOutside0, feeGrowthOutside1
````

````python
# snapshot in time
curr_tick = int(df['tick'].values[-1])
feeGrowthGlobal_X = df['acc_fees_X'].values[-1]
feeGrowthGlobal_Y = df['acc_fees_Y'].values[-1]

# TODO we need to know the tick spacing delta_pi for this POOL!
rows=[]
for tau in np.arange(np.round(df['tick'].min(), decimals=-1), np.round(df['tick'].max(), decimals=-1), 10):
    # unitary range [tau, tau+1)
    
    feeGrowthOutside_lower_X, feeGrowthOutside_lower_Y = getFeeGrowthOutside(tau)
    feeGrowthOutside_upper_X, feeGrowthOutside_upper_Y = getFeeGrowthOutside(tau+1)
    
    if(curr_tick >= tau):
        feeBelow_X = feeGrowthOutside_lower_X
        feeBelow_Y = feeGrowthOutside_lower_Y
    else:
        feeBelow_X = feeGrowthGlobal_X - feeGrowthOutside_lower_X
        feeBelow_Y = feeGrowthGlobal_Y - feeGrowthOutside_lower_Y
    
    if (curr_tick >= tau+1):
        feeAbove_X = feeGrowthGlobal_X - feeGrowthOutside_upper_X
        feeAbove_Y = feeGrowthGlobal_Y - feeGrowthOutside_upper_Y
        
    else:
        feeAbove_X = feeGrowthOutside_upper_X
        feeAbove_Y = feeGrowthOutside_upper_Y

    fees_R_X = feeGrowthGlobal_X - feeBelow_X - feeAbove_X
    fees_R_Y = feeGrowthGlobal_Y - feeBelow_Y - feeAbove_Y

    rows.append({'fees_R_X':fees_R_X, 'fees_R_Y': fees_R_Y})

df3 = pd.DataFrame(rows)
````

````python
rows = []
with open("data/logs.txt") as f:
    for line in f.readlines():
        sp1 = line.split(":")
        rows.append({
            'tau': int(float(sp1[0].strip())),
            'fees_R_X': Decimal(sp1[1].split(",")[0].strip()),
            'fees_R_Y': Decimal(sp1[1].split(",")[1].strip())
        })
df4 = pd.DataFrame(rows)
df4['price'] = 1/((beta_p**df4['tau'].apply(Decimal)) / Decimal(10**(WETH_dec-USDC_dec)))
````

````python
rows1 = []
for tau in np.arange(np.round(df['tick'].min(), decimals=-1), np.round(df['tick'].max(), decimals=-1), 10):
    tmp = df4.query(f'tau == {tau}')
    if(tmp.shape[0] == 0):
        rows1.append({
            'tau': int(tau),
            'fees_R_X': 0,
            'fees_R_Y': 0,
        })
        continue
        
    rows1.append({
        'tau': int(tau),
        'fees_R_X': tmp.iloc[0]['fees_R_X'],
        'fees_R_Y': tmp.iloc[0]['fees_R_Y'],
    })
df5 = pd.DataFrame(rows1)
````

````python
N = df.shape[0]
for i in range(N):
    row = df.iloc[i]
    query1 = f"""
    {{
      ticks(where: {{ 
        	pool: "{WETH_USDC_pool_addr}", 
        	tickIdx: {int(row['tick'])} 
      }}) {{
        feeGrowthOutside0X128
        feeGrowthOutside1X128
      }}
    }}
    """
    rows1=[]
    raw_data1 = send_graphql_query_to_subgraph(api_keys['SUBGRAPH_API_KEY'], query1)
    for ele in raw_data1["data"]["ticks"]:
        if(ele["feeGrowthOutside0X128"] =="0" or ele["feeGrowthOutside1X128"] =="0"):
            continue
        rows1.append({
            "fees_out_X": ele["feeGrowthOutside0X128"],
            "fees_out_Y": ele["feeGrowthOutside1X128"]
        })
    df1 = pd.DataFrame(rows1)
    acc_gt_X = (df1['fees_out_X'].astype("str").apply(Decimal) / Q128).sum()
    acc_gt_Y = (df1['fees_out_Y'].astype("str").apply(Decimal) / Q128).sum()
    
    query2 = f"""
    {{
      ticks(where: {{ 
        	pool: "{WETH_USDC_pool_addr}", 
        	tickIdx_lt: {int(row['tick'])} 
      }}) {{
        feeGrowthOutside0X128
        feeGrowthOutside1X128
      }}
    }}
    """
    
    rows2=[]
    raw_data2 = send_graphql_query_to_subgraph(api_keys['SUBGRAPH_API_KEY'], query2)
    for ele in raw_data2["data"]["ticks"]:
        if(ele["feeGrowthOutside0X128"] =="0" or ele["feeGrowthOutside1X128"] =="0"):
            continue
        rows2.append({
            "fees_out_X": ele["feeGrowthOutside0X128"],
            "fees_out_Y": ele["feeGrowthOutside1X128"]
        })
    df2 = pd.DataFrame(rows2)
    acc_lt_X = (df2['fees_out_X'].astype("str").apply(Decimal) / Q128).sum()
    acc_lt_Y = (df2['fees_out_Y'].astype("str").apply(Decimal) / Q128).sum()

    df['fees_X_R'] = df['acc_fees_X'] - acc_gt_X - acc_lt_X
    df['fees_Y_R'] = df['acc_fees_Y'] - acc_gt_Y - acc_lt_Y
````

````python
fig, ax = plt.subplots(1,1,figsize=(12,4))
sns.lineplot(data=df, x="price", y="fees_X_R", ax=ax)
plt.show()
````
